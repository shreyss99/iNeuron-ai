```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diabetes = pd.read_csv("diabetes.txt", sep=",")
```

```python
diabetes
```

```python
diabetes.describe()
```

```python
x = diabetes.drop(columns='Outcome')
```

```python
x
```

```python
y = diabetes[['Outcome']]
```

```python
y
```

```python
scaler = StandardScaler()
x_scaled_arr = scaler.fit_transform(x)
```

```python
x_scaled_arr
```

```python
x_scaled_df = pd.DataFrame(x_scaled_arr)
```

```python
x_scaled_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=30, test_size=0.25)
```

```python
naive_bayes = GaussianNB()
```

```python
naive_bayes.fit(x_train, y_train)
```

```python
naive_bayes.predict(x_test)
```

```python
naive_bayes.score(x_test, y_test)
```
