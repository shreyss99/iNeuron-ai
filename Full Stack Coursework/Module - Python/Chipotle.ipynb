```python
# Total quantity
sum(chipotle['quantity'])
```

```python
# total item price
sum = 0
for i in range(len(chipotle['item_price'])):
    sum += float((chipotle['item_price'][i][1:]).strip())
    
sum
```

```python
# total item price * quantity

sum = 0
for i in range(len(chipotle['item_price'])):
    sum += float((chipotle['item_price'][i][1:]).strip()) * chipotle['quantity'][i]
    
sum
```

```python
# count of items ordered by group

x = chipotle.groupby('item_name')
x = x.sum()
x = x.sort_values(['quantity'])
x.head(1000)
```

```python
# max items ordered 

x = chipotle.groupby('item_name')
x = x.sum()
x = x.sort_values(['quantity'], ascending=False)
x.head(1)
```
