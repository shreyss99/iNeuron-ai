```python
'''
Question 1:
Define a class with a generator which can iterate the numbers, which are divisible by
7, between a given range 0 and n.
'''

class Divisible:
    
    def __init__(self,n):
        self.n = n
        
    def generator(self):
        for i in range(1,self.n):
            if i % 7 == 0:
                yield i

n = int(input("Enter the limit for n: "))
divisible_7 = Divisible(n)
for i in divisible_7.generator():
    print(i)
```

```python
'''
Question 2:
Write a program to compute the frequency of the words from the input. The output
should output after sorting the key alphanumerically.
Suppose the following input is supplied to the program:
New to Python or choosing between Python 2 and Python 3? Read Python 2 or Python 3.
Then, the output should be:
2:2
3.:1
3?:1
New:1
Python:5
Read:1
and:1
between:1
choosing:1
or:2
to:1
'''

def count_freq(string):
    list = string.split(" ")
    set_single_words = set(list)
    
    sorted_list = []
    for i in set_single_words: 
        sorted_list.append(i)
    sorted_list.sort()
    
    count_list = []
    for i in sorted_list:
        count_list.append([i,list.count(i)])
    return count_list
    
string = input("Enter the string: ")
result_list = count_freq(string)

print()

for i in result_list:
    print(i[0]+":"+str(i[1]))
```

```python
'''
Question 3:

Define a class Person and its two child classes: Male and Female. All classes have a
method "getGender" which can print "Male" for Male class and "Female" for Female
class.
'''

class Person:  
    def getGender(self):
        print("Unknown")
    
class Male(Person):
    def getGender(self):
        print("Male")
        
class Female(Person):
    def getGender(self):
        print("Female")
        
male = Male()
female = Female()

male.getGender()
female.getGender()
```

```python
'''
Question 5:
Please write a program to compress and decompress the string 
"hello world!hello world!hello world!hello world!".
'''

import gzip
string = b'hello world!hello world!hello world!hello world!'
compress = gzip.compress(string)
print("Compressed: ",compress)
decompress = gzip.decompress(compress)
print("Decompressed: ",decompress)
```

```python
'''
Question 6:
Please write a binary search function which searches an item in a sorted list. The
function should return the index of element to be searched in the list.
'''

def binary_search_index(list,low,high,search_val):
    list.sort()
    mid = (low + high) // 2
    
    while(low <= high):
        if search_val == list[mid]:
            return mid
        elif search_val < list[mid]:
            return binary_search_index(list,low,mid,search_val)
        else:
            return binary_search_index(list,mid+1,high,search_val)

length = int(input("Enter the number of elements in the list: "))
list = []
for i in range(length):
    list.append(int(input()))
search_val = int(input("Enter the element to be searched for: "))

low = 0
high = len(list) - 1
index_loc = binary_search_index(list,low,high,search_val)
print("The element {} is preset at index {} in the list".format(search_val,index_loc))
```
