```python
'''
Q1. You are given two strings s and t. String t is generated by randomly
shuffling string s and then adding one more letter at a random position.
Return the letter that was added to t.
Examples
find_the_difference("abcd", "abcde") ➞ "e"
find_the_difference("", "y") ➞ "y"
find_the_difference("ae", "aea") ➞ "a"
'''

def find_the_difference(s,t):
    list_s = []
    list_t = []
    difference_flag = False
    for i in s:
        list_s.append(i)
    for i in t:
        list_t.append(i)
    list_s.sort()
    list_t.sort()
    for i in range(len(s)):
        if list_s[i] != list_t[i]:
            difference_flag = True
            break
        else:
            continue
        
    if difference_flag == True:
        difference_char = list_t[i]
    elif difference_flag == False:
        difference_char = list_t[len(list_t) - 1]
    
    return difference_char
    

s = input("Enter the string s: ")
t = input("Enter the string t: ")
print("The difference between strings s and t is {}".format(find_the_difference(s,t)))
```

```python
'''
Q2. Given a function that accepts unlimited arguments, check and count how
many data types are in those arguments. Finally return the total in a list.
List order is:
[int, str, bool, list, tuple, dictionary]
Examples
count_datatypes(1, 45, "Hi", False) ➞ [2, 1, 1, 0, 0, 0]
count_datatypes([10, 20], ("t", "Ok"), 2, 3, 1) ➞ [3, 0, 0, 1, 1, 0]
count_datatypes("Hello", "Bye", True, True, False, {"1": "One", "2": "Two"}, [1,
3], {"Brayan": 18}, 25, 23) ➞ [2, 2, 3, 1, 0, 2]
count_datatypes(4, 21, ("ES", "EN"), ("a", "b"), False, [1, 2, 3], [4, 5, 6]) ➞ [2,
0, 1, 2, 2, 0]
'''

def count_datatypes(*args):
    count_int = 0
    count_str = 0
    count_bool = 0
    count_list = 0
    count_tuple = 0
    count_dict = 0
    for i in args:
        if type(i) == int:
            count_int += 1
        elif type(i) == str:
            count_str += 1
        elif type(i) == bool:
            count_bool += 1
        elif type(i) == list:
            count_list += 1
        elif type(i) == tuple:
            count_tuple += 1
        elif type(i) == dict:
            count_dict += 1
    
    return [count_int,count_str,count_bool,count_list,count_tuple,count_dict]

count_datatypes(4, 21, ("ES", "EN"),True, ("a", "b"), {"a":23,"code":"works"}, False, [1, 2, 3], [4, 5, 6], "full_stack", 45,
               ['a','b'])
```

```python
'''
Q3. A Fibonacci string is a precedence of the Fibonacci series. It works with
any two characters of the English alphabet (as opposed to the numbers 0 and
1 in the Fibonacci series) as the initial items and concatenates them together
as it progresses in a similar fashion as the Fibonacci series.
Examples
fib_str(3, ["j", "h"]) ➞ "j, h, hj"
fib_str(5, ["e", "a"]) ➞ "e, a, ae, aea, aeaae"
fib_str(6, ["n", "k"]) ➞ "n, k, kn, knk, knkkn, knkknknk"
'''

def fib_str(no_of_values, string_list):
    series = []
    first = string_list[0]
    second = string_list[1]
    series.append(first)
    series.append(second)
    for i in range(no_of_values - 2):
        next = second + first
        series.append(next)
        second,first = next,second
    return series
    
no_of_values = int(input("Enter the number of values need in final Fibonacci string series: "))
string_list = []
for i in range(2):
    string_list.append(input("Enter the character for developing Fibonacci string series: "))

print("The resulting Fibonacci string series is {}".format(fib_str(no_of_values,string_list)))
```

```python
'''
Q4. Given an integer between 0 and 26, make a variable (self.answer). That
variable would be assigned to a string in this format:
"nines:your answer, threes:your answer, ones:your answer"
You need to find out how many ones, threes, and nines it would at least take
for the number of each to add up to the given integer when multiplied by one,
three or nine (depends).
Examples
ones_threes_nines(10) ➞ "nines:1, threes:0, ones:1"
ones_threes_nines(15) ➞ "nines:1, threes:2, ones:0"
ones_threes_nines(22) ➞ "nines:2, threes:1, ones:1"
'''

def ones_threes_nines(number):
    count_9 = 0
    count_3 = 0
    count_1 = 0
    while number != 0:
        while number >= 9:
            count_9 += 1
            number -= 9
        while number >= 3:
            count_3 += 1
            number -= 3
        while number >= 1:
            count_1 += 1
            number -= 1
    print("nines:{}, threes:{}, ones:{}".format(count_9,count_3,count_1))
    
number = int(input("Enter the number: "))
ones_threes_nines(number)
```

```python
'''
Q5. The Fibonacci sequence is a classic use case for recursive functions since
the value of the sequence at a given index is dependent on the sum of the
last two values. However, the recursion tree created by solving the Fibonacci
sequence recursively can grow quite fast. Therefore it can be important to
think about the implications of running a function recursively. Depending on
the size of n needed and the capabilities of the system in question you might
want to take a different approach.
Write a non-recursive function that takes an integer n and returns the
Fibonacci sequence's value at index n.
Examples
fib(6) ➞ 8
# 0 + 1 = 1, 1 + 1 = 2, 1 + 2 = 3, 2 + 3 = 5, 3 + 5 = 8
fib(1) ➞ 0
fib(2) ➞ 1
'''

def fib(number):
    a = 0
    b = 1
    for i in range(number):
        a,b = b,a+b
    return a

number = int(input("Enter the index for Fibonacci series: "))
index_value = fib(number)
print("The value of Fibonacci series at index {} is {}".format(number,index_value))
```
