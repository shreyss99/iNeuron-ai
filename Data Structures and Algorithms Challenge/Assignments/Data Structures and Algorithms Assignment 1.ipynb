```python
'''
Q1. Describe Python's built-in data structure?

Ans) Python has 4 major built-in data structures: List, Tuple, Set and Dictionary.

A) LIST:

- It is a data structure that holds an ordered collection of items in a sequential manner. 
- You define a list by placing elements inside a [] separated by comma.
- There are addresses assigned to every element called as index. The index value starts from 0 and goes till last element.
- It is possible to store elements of different data types inside a list as opposed to array which restricts the data type
  must be same for all elements. It means we can store integers, floats, strings, complex numbers, lists etc inside a list.
- Another unique feature is that lists are mutable which means that the elements inside a list and be changed/updated.
- There are many operations you can perform - add and remove elements, sort, find length and index of elements etc.
- We can perform slicing and concatenaton of list elements as well.

B) TUPLE:

- Similar to a list, a tuple is another built-in data type but differs in two things. 
- Firstly, tuples are immutable; once you define values in a tuple, you cannot make any changes. 
- Secondly, we use () to define the set of values in a tuple.
- There are addresses assigned to every element called as index. The index value starts from 0 and goes till last element.
- It is possible to store elements of different data types inside a tuple.
- A tuple with 0 or 1 element is uniquely specified with element followed by a comma within ().
- There are many operations you can perform - find length and index of elements etc.
- Adding or removing elements from tuple is not possible as it is immutable.

C) SET:

- A set comprises of unique and unordered elements. 
- Even if the element is present more than once, it will be counted once in the set list. 
- We use {} to define elements in a set.
- Set elements cannot be indexed via []. They are not subscriptable.
- It is possible to store elements of different data types inside a set.
- There are many operations you can perform - add and remove elements, sort, find length and index of elements etc.
- There are special set operations like unin, intersection, difference and symmetric difference we can perform.
- We cannot concatenate or replicate a set.

D) DICTIONARY:

- Dictionaries are used to store key-value pairs also declared in {}.
- We associate keys (name) with values (details). 
- Key must be unique because you cannot find the correct information if you have two elements with the exact same key.
- We can use only immutable objects for the keys of a dictionary but can use either immutable or mutable objects 
  for the values of the dictionary.
- Pairs of keys and values are specified in a dictionary by using the notation d = {key1 : value1, key2 : value2 }. 
  Key-value pairs are separated by a colon and the pairs are separated themselves by commas.
- We can get all the keys and values of the dictionary in a list using values() and keys() functions.

'''


print("LIST: ")
my_List = [1, 2.0, "abc", 4-7j] 
print(f"Original list -- {my_List}")
my_List.append([3,4,5])
print(f"List after appending '[3,4,5]' -- {my_List}")
my_List.remove(4-7j)
print(f"List after removing '4-7j' -- {my_List}")
print(f"Length of list -- {len(my_List)}")
print(f"Index of 'abc' -- {my_List.index('abc')}")
print(f"Concatenate the same list  -- {my_List + my_List}")
print(f"Slicing elements from index 2 to 4 -- {my_List[2:4]}")

print()
print("TUPLE: ")
my_Tuple = ('pqr', -5.0, "shrey", (4,5,6), 99)
print(f"Original tuple -- {my_Tuple}")
print(f"Length of tuple -- {len(my_Tuple)}")
print(f"Index of '99' -- {my_Tuple.index(99)}")
print(f"Concatenate the same tuple  -- {my_Tuple + my_Tuple}")
print(f"Slicing elements from index 1 to -1 -- {my_Tuple[1:-1]}")
print(f"Empty tuple -- {tuple()}")
print(f"Tuple with 1 element -- {(23, )}")

print()
print("SET: ")
my_Set = {'def', -5.0, (4,5,6), 99, '', 'def', -5.0, "shrey", (3-4j), 99, ''}
print(f"Original set -- {my_Set}")
print(f"Length of set before adding -5.0 -- {len(my_Set)}")
print(f"Adding -5.0 to the set -- {my_Set.add(-5.0)}")
print(f"Length of set after adding -5.0 -- {len(my_Set)}")
print(f"Removing '(4,5,6)' from the set -- {my_Set.remove((4,5,6))}")
print(f"Set contents -- {my_Set}")
print("Set Operations")
my_set = {1, 2, 3, 4}
my_set_2 = {3, 4, 5, 6}
print(my_set)
print(my_set_2)
print(f"Union -- {my_set.union(my_set_2)}")
print(f"Intersection -- {my_set.intersection(my_set_2)}")
print(f"Difference -- {my_set.difference(my_set_2)}")
print(f"Symettric Difference -- {my_set.symmetric_difference(my_set_2)}")

print()
print("DICTIONARY: ")
my_Dict = {
    'Swaroop': 'swaroop@swaroopch.com',
    'Larry': 'larry@wall.org',
    'Matsumoto': 'matz@ruby-lang.org',
    'Spammer': 'spammer@hotmail.com'
}
print(f"Original dictionary -- {my_Dict}")
print(f"Length of dictionary -- {len(my_Dict)}")
print(f"Larry's address is -- {my_Dict['Larry']}")
print(f"Deleting spammer address")
del my_Dict['Spammer']
print(f"Dictionary contents -- {my_Dict}")
my_Dict['Guido'] = 'guido@python.org'
print(f"Adding a new address -- ")
print(f"Dictionary contents -- {my_Dict}")

salary = {"raj" : 50000, "striver" : 60000, "vikram" : 5000} 
print(f"Salary Dictionary -- {salary}")
print(f"List of values -- {salary.values()}")
print(f"List of keys -- {salary.keys()}")
```

```python
'''
Q2. Describe the Python user data structure?

Ans) Below are the user-defined data structures in Python:

A) ARRAYS:

- Arrays and lists are the same structure with 1 difference. 
- Lists allow heterogeneous data element storage whereas arrays allow only homogeneous elements to be stored within them.

B) STACK:

- Stacks are linear Data Structures based on the principle of Last-In-First-Out (LIFO).
- Data which is entered last will be the first to get accessed. 
- It is built using the array structure and has operations namely, pushing (adding) elements, popping (deleting) elements and     accessing elements only from one point in the stack called as the Top. 
- This Top is the pointer to the current position of the stack. 
- There are methods to check if a stack is empty or full.
- If a stack is full and we try to add an element it results in an overflow condition.
- Similarly, if a stack is empty and we try to remove an element it results in an underflow condition.
- Stacks are prominently used in applications such as Recursive Programming, check well-formedness of parantheses,
  reversing words, storing addresses of function calls in hierarchical order.
  
C) QUEUE:

- A queue is a linear data structure which is based on the principle of First-In-First-Out (FIFO).
- Data entered first will be accessed first.
- It is built using the array structure and has operations which can be performed from both ends of the Queue, head- tail or 
  front-back.
- The end where elements are inserted is clled the Front and the end from where they are removed is called Rear.
- Operations such as adding and deleting elements are called En-Queue and De-Queue and accessing the elements can be performed.
- There are methods to check if a stack is empty or full.
- There are circular queues as well where the front and rear join each other to form a circle. 
- Queues are used as Network Buffers for traffic congestion management, used in Operating Systems for Job Scheduling and many more.

D) LINKED LIST:

- Linked lists are linear Data Structures which are not stored contigiously in memory.
- The linked list consists a series of elements called as nodes linked with each other.
- The node of a linked list is composed of data and a pointer called next. 
- The next pointer determines which is the next node in the list. 
- There is a Head pointer which points to the first node of the linked list. 
- The last node of the linked list does not point to any other node and its pointer stores Null.
- We can perform addition and removal of elements on linked list.
- There are singly linked-list where there is only single direction of connection to the next node. Doubly linked-list are
  those where we have both directions and we have 2 pointers to the next and previous node.
- These structures are most widely used in image viewing applications and so forth.

E) TREES:

- Trees are non-linear Data Structures which have root and nodes. 
- The root is the node from where the data originates and the nodes are the other data points that are available to us. 
- The node that precedes is the parent and the node after is called the child. 
- There are levels a tree has to show the depth of information. 
- The last nodes are called the leaves.
- There are various methods for tree traversal - pre-order, in-order and post-order traversals.
- We have multiple types of trees, binary tree, binary serch tree, AVL trees, Red-Black trees etc.
- Trees create a hierarchy which can be used in a lot of real-world applications such as the HTML pages use trees to distinguish   which tag comes under which block. It is also efficient in searching purposes and rule based classification.

F) GRAPHS:

- Graphs are non-linear data structure used to store data collection of points.
- They are stored in form of vertices (nodes) and edges (edges). 
- The edges may be directional or un-directional and based on this we have directed and undirected graphs. Indirected, we 
  need to follow the edge direction. We can also have bidirectional graphs. Complete graphs are those in which each vertex
  is connected to other vie an edge.
- Graphs can be called as the most accurate representation of a real-world map. 
- We have various graph traversal algorithms - Breadth First Search and Depth First Search traversals.
- They are used to find the various cost-to-distance between the various data points called as the nodes and hence find the       least path. Many applications such as Google Maps, Uber, and many more use Graphs to find the least distance and increase       profits in the best ways.

G) HASHMAP:

- Hash Maps are indexed data structures and perform the same function as that performed by dictionary in Python. 
- A hash map uses a hash function to compute index-key values into arrays. 
- Hashmap is a type of data structure that maps keys to its value pairs.
- It basically makes use of a function that computes an index value that in turn holds the elements to be searched, inserted, removed, etc. 
- This makes it easy and fast to access data. 
- In general, hash tables store key-value pairs and the key is generated using a hash function.

'''
```
