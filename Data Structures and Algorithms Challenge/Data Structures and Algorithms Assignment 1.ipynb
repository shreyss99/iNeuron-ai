```python
'''
Q1. Describe Python's built-in data structure?

Ans) Python has 4 major built-in data structures: List, Tuple, Set and Dictionary.

A) LIST:

- It is a data structure that holds an ordered collection of items in a sequential manner. 
- You define a list by placing elements inside a [] separated by comma.
- There are addresses assigned to every element called as index. The index value starts from 0 and goes till last element.
- It is possible to store elements of different data types inside a list as opposed to array which restricts the data type
  must be same for all elements. It means we can store integers, floats, strings, complex numbers, lists etc inside a list.
- Another unique feature is that lists are mutable which means that the elements inside a list and be changed/updated.
- There are many operations you can perform - add and remove elements, sort, find length and index of elements etc.
- We can perform slicing and concatenaton of list elements as well.

B) TUPLE:

- Similar to a list, a tuple is another built-in data type but differs in two things. 
- Firstly, tuples are immutable; once you define values in a tuple, you cannot make any changes. 
- Secondly, we use () to define the set of values in a tuple.
- There are addresses assigned to every element called as index. The index value starts from 0 and goes till last element.
- It is possible to store elements of different data types inside a tuple.
- A tuple with 0 or 1 element is uniquely specified with element followed by a comma within ().
- There are many operations you can perform - find length and index of elements etc.
- Adding or removing elements from tuple is not possible as it is immutable.

C) SET:

- A set comprises of unique and unordered elements. 
- Even if the element is present more than once, it will be counted once in the set list. 
- We use {} to define elements in a set.
- Set elements cannot be indexed via []. They are not subscriptable.
- It is possible to store elements of different data types inside a set.
- There are many operations you can perform - add and remove elements, sort, find length and index of elements etc.
- There are special set operations like unin, intersection, difference and symmetric difference we can perform.
- We cannot concatenate or replicate a set.

D) DICTIONARY:

- Dictionaries are used to store key-value pairs also declared in {}.
- We associate keys (name) with values (details). 
- Key must be unique because you cannot find the correct information if you have two elements with the exact same key.
- We can use only immutable objects for the keys of a dictionary but can use either immutable or mutable objects 
  for the values of the dictionary.
- Pairs of keys and values are specified in a dictionary by using the notation d = {key1 : value1, key2 : value2 }. 
  Key-value pairs are separated by a colon and the pairs are separated themselves by commas.
- There are many operations you can perform - add and remove elements, sort, find length and index of elements etc.
- There are special set operations like unin, intersection, difference and symmetric difference we can perform.
- We cannot concatenate or replicate a set.

'''


print("LIST: ")
my_List = [1, 2.0, "abc", 4-7j] 
print(f"Original list -- {my_List}")
my_List.append([3,4,5])
print(f"List after appending '[3,4,5]' -- {my_List}")
my_List.remove(4-7j)
print(f"List after removing '4-7j' -- {my_List}")
print(f"Length of list -- {len(my_List)}")
print(f"Index of 'abc' -- {my_List.index('abc')}")
print(f"Concatenate the same list  -- {my_List + my_List}")
print(f"Slicing elements from index 2 to 4 -- {my_List[2:4]}")

print()
print("TUPLE: ")
my_Tuple = ('pqr', -5.0, "shrey", (4,5,6), 99)
print(f"Original tuple -- {my_Tuple}")
print(f"Length of tuple -- {len(my_Tuple)}")
print(f"Index of '99' -- {my_Tuple.index(99)}")
print(f"Concatenate the same tuple  -- {my_Tuple + my_Tuple}")
print(f"Slicing elements from index 1 to -1 -- {my_Tuple[1:-1]}")
print(f"Empty tuple -- {tuple()}")
print(f"Tuple with 1 element -- {(23, )}")

print()
print("SET: ")
my_Set = {'def', -5.0, (4,5,6), 99, '', 'def', -5.0, "shrey", (3-4j), 99, ''}
print(f"Original set -- {my_Set}")
print(f"Length of set before adding -5.0 -- {len(my_Set)}")
print(f"Adding -5.0 to the set -- {my_Set.add(-5.0)}")
print(f"Length of set after adding -5.0 -- {len(my_Set)}")
print(f"Removing '(4,5,6)' from the set -- {my_Set.remove((4,5,6))}")
print(f"Set contents -- {my_Set}")
print("Set Operations")
my_set = {1, 2, 3, 4}
my_set_2 = {3, 4, 5, 6}
print(my_set)
print(my_set_2)
print(f"Union -- {my_set.union(my_set_2)}")
print(f"Intersection -- {my_set.intersection(my_set_2)}")
print(f"Difference -- {my_set.difference(my_set_2)}")
print(f"Symettric Difference -- {my_set.symmetric_difference(my_set_2)}")

print()
print("DICTIONARY: ")
my_Dict = {
    'Swaroop': 'swaroop@swaroopch.com',
    'Larry': 'larry@wall.org',
    'Matsumoto': 'matz@ruby-lang.org',
    'Spammer': 'spammer@hotmail.com'
}
print(f"Original dictionary -- {my_Dict}")
print(f"Length of dictionary -- {len(my_Dict)}")
print(f"Larry's address is -- {my_Dict['Larry']}")
print(f"Deleting spammer address")
del my_Dict['Spammer']
print(f"Dictionary contents -- {my_Dict}")
my_Dict['Guido'] = 'guido@python.org'
print(f"Adding a new address -- ")
print(f"Dictionary contents -- {my_Dict}")
```
