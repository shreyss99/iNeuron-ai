### Q1. What is the divide and conquer strategy?

Ans)

In the divide and conquer approach, firstly a problem is divided into smaller problems, then these smaller problems are solved independently, and then finally the solutions of these smaller problems are combined or merged in order to get a solution for the large or the main problem. 

The divide and Conquer algorithm can be divided into the following three steps.

- Divide: Divide the original problem into a set of smaller problems or subproblems.
- Conquer: Solve every subproblem individually and repeat it i.e., recursively.
- Combine: We put together or merge the solutions of the subproblems in order to get the solution for the whole problem.

Fundamentals of the Divide and Conquer Strategy:
There are two fundamentals of the Divide and Conquer Strategy. Let us a look at them.
 
- Relational Formula: This is something that we generate from the given technique. Once the formula is generated, we apply the Divide and Conquer Strategy.
- Stopping Condition: Whenever we break the problem, we must have an idea till when will our program run and what should be the end. So, wherever the base condition of recursion occurs, that condition is called a Stopping Condition.

Advantages of Divide and Conquer:
- It becomes easy for us to solve the biggest problems. We can lessen the effort that it used to take as it works mainly on dividing the problem into two halves and then solving them recursively.
- The memory is efficiently used by the strategy as it does not occupy much space and solves simple subproblems within the memory rather than accessing the slower main memory.
- It is much more proficient than the Brute Force technique which is its counterpart.

Disadvantages of Divide and Conquer:
- As most of its algorithms are designed by incorporating recursion, it might require high memory management for proper functioning.
- An explicit stack might lead to overusing the space.
- There is a chance for the system to crash if the recursion is performed continuously - if it exceeds the stack limit.

Examples of Problems solved using Divide and Conquer:
- Binary Searching
- Merge Sort
- Quick Sort
- Finding maximum and minimum numbers in an array
- Strassen's matrix multiplication.

### Q3. Explain the distinction between a list and a tuple.

Ans) 

A) LIST:

- It is a data structure that holds an ordered collection of items in a sequential manner. 
- You define a list by placing elements inside a [] separated by comma.
- There are addresses assigned to every element called as index. The index value starts from 0 and goes till last element.
- It is possible to store elements of different data types inside a list as opposed to array which restricts the data type
  must be same for all elements. It means we can store integers, floats, strings, complex numbers, lists etc inside a list.
- Another unique feature is that lists are mutable which means that the elements inside a list and be changed/updated.
- There are many operations you can perform - add and remove elements, sort, find length and index of elements etc.
- We can perform slicing and concatenaton of list elements as well.
- The implication of iterations is time-consuming in the list.
- Operations like insertion and deletion are better performed.
- Consumes more memory.
- Many built-in methods are available.
- Unexpected errors and changes can easily occur in lists.

B) TUPLE:

- Similar to a list, a tuple is another built-in data type but differs in two things. 
- Firstly, tuples are immutable; once you define values in a tuple, you cannot make any changes. 
- Secondly, we use () to define the set of values in a tuple.
- There are addresses assigned to every element called as index. The index value starts from 0 and goes till last element.
- It is possible to store elements of different data types inside a tuple.
- A tuple with 0 or 1 element is uniquely specified with element followed by a comma within ().
- There are many operations you can perform - find length and index of elements etc.
- Adding or removing elements from tuple is not possible as it is immutable.
- Implications of iterations are much faster in tuples.
- Elements can be accessed better.
- Consumes less memory.
- Does not have many built-in methods.
- Unexpected errors and changes rarely occur in tuples.

### Q4. Can you explain how Python manages memory?

Ans)

- Memory Management is the process of storing memory dynamically and freeing it when not in use.
- In Python, this allocation and deallocation of memory are done automatically by the Python garbage collector created by python developers so that users don’t have to do manual garbage collection.
- Garbage collection in Python is the memory management process of freeing up the unused and unwanted memory for our programs by the interpreter. In Python, this is done automatically.
- Iy involves a private heap containing all Python objects and data structures. Interpreter takes care of Python heap and that the programmer has no access to it.
- The allocation of heap space for Python objects is done by Python memory manager. The core API of Python provides some tools for the programmer to code reliable and more robust program.
- Python also has a build-in garbage collector which recycles all the unused memory. When an object is no longer referenced by the program, the heap space it occupies can be freed. 
- Reference counting is a technique in which an object is deallocated from the memory when there are no references pointing to it. When the reference count becomes zero, the object is deleted.
-  The gc module defines functions to enable /disable garbage collector:
   - gc.enable() - Enables automatic garbage collection.
   - gc.disable() - Disables automatic garbage collection.

### Q5. What is the difference between pickling and unpickling?

Ans)

**Pickling:**

Pickle is used for serializing and de-serializing Python object structures, also called marshalling or flattening. Serialization refers to the process of converting an object in memory to a byte stream that can be stored on disk or sent over a network. 

Pickling is useful for applications where you need some degree of persistency in your data. Your program's state data can be saved to disk, so you can continue working on it later on. It can also be used to store python objects in a database. Pickle is very useful for when you're working with machine learning algorithms, where you want to save them to be able to make new predictions at a later time, without having to rewrite everything or train the model all over again.

**Unpickling:**

Unpickling is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy. It is used in deserialzation of byte stream to object in memory. The character stream can then be retrieved and de-serialized back to a Python object.
